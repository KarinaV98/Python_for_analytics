### 2.1. Изучите распределение домов от наличия вида на набережную

```python
df.head()
```

```python
data_1 = df['view'].value_counts()
names = data_1.index
values = data_1.values
```

```python
df['view'].value_counts()
```

```python
def calcExplodes(data):
    exp = []
    k = 1
    for val in data.values:
        if val < data.values.sum() / 20:
            exp.append(0.3 * k)
            k += 1
        else:
            exp.append(0)
    return exp
```

```python
data_1 = df['waterfront'].value_counts()
data_1.index = ['Без видa','С видом']

plt.pie(data_1.values, labels=data_1.index, autopct='%1.1f%%', pctdistance=0.8,
        explode=calcExplodes(data_1))

plt.title("Распределение домов в зависимости от наличия вида на набережную")
```

На продажу выставлено менее 1% домов с видом на набережную.

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['sqft_living'], y=df['price'],  color = '#778899')

plt.title("Зависимость цены от размера площади")
plt.xlabel("Жилая площадь")
plt.ylabel("Цена дома, млн");
```

Одним из основных критериев повышения цен на недвижимость является жилая площадь. Чем больше квадратура - тем выше цена.

```python
sns.barplot(x=df['condition'], y=df['price'], color = '#00FF00')

plt.title("Зависимость цены от состояния дома")
plt.xlabel("Оценка состояния дома")
plt.ylabel("Средняя цена дома");
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 9))

sns.scatterplot(x=df['bathrooms'], y=df['price'], color = '#00FF00')
plt.title('Зависимость цены от кол-ва ванных комнат')
plt.xlabel('bathrooms')
plt.show()
```

Наблюдается рост цен в зависимости от кол-ва ванных комнат. Однако, есть незначительные выбросы, указывающие на то, что кол-во ванных комнат не всегда влияет на окончательную стоимость.

```python
plt.figure(figsize=(10, 9))

sns.lineplot(x=df['bedrooms'], y=df['price'], color = 'purple')
plt.title('Зависимость цены от кол-ва спален')
plt.xlabel('bedrooms')
```

Наиболее высокую стоимость имеюд дома с 6-8 спальнями. Дома, в которых более 8 спален резко падают в цене. Возможно, это связано с типом жилья (дома на несколько семей).

```python
sns.barplot(x=df['floors'], y=df['price'], color = '#00CED1')
plt.title('Зависимость цены от кол-ва этажей')
plt.xlabel('floors')
```

Самые недорогими являются одноэтажные дома. Даиболее высокую стоимость имеют дома в 2,5 этажей.
