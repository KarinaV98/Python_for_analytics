### 1.2. Изучите стоимости недвижимости

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./kc-house-data (1).csv')
```

```python
df
```

```python
df['price'].nunique()
```

```python
plt.figure(figsize=(8,3))
sns.histplot(df['price'], bins=100)
plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество домов на продажу');
```

Большая часть ноутбуков стоит примерно 

### 1.3. Изучите распределение квадратуры жилой площади

```python
df.head()
```

```python
df['sqft_living'].value_counts()
```

```python
plt.figure(figsize=(8,3))
sns.histplot(df['sqft_living'], bins=20)
plt.title('Квадратура жилой площади')
plt.xlabel('Квадратура')
plt.ylabel('Количество домов на продажу');
```

Большая часть домов, выставленных на продажу, имеет пложадь примерно 1500 - 2100 квадратных метров. Домов с площадью от 6000 до 8000 квадратных метров меньше всего.

### 1.4. Изучите распределение года постройки

```python
df['yr_built'].value_counts()
```

```python
plt.figure(figsize=(8,3))
sns.histplot(df['yr_built'], bins=20)
plt.title('Распределение года постройки')
plt.xlabel('Год')
plt.ylabel('Количество домов');
```

### 2.1. Изучите распределение домов от наличия вида на набережную

```python
df.head()
```

```python
data_1 = df['view'].value_counts()
names = data_1.index
values = data_1.values
```

```python
df['view'].value_counts()
```

```python
plt.figure(figsize=(6,6))

plt.pie(values, autopct = '%1.1f%%', labels = names);
plt.title('Распределение домов от наличия вида на набережную')
plt.legend(names, loc = 'best');
```
